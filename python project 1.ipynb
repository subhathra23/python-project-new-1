```python
#import libraries to create visulization
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#creating a new dataFrame with Year and sales data
data = {
    'Year': [2010, 2011, 2012, 2013, 2014, 2015, 2016],
    'Sales': [500, 800, 1200, 900, 1500, 1700, 2000]
}
df = pd.DataFrame(data)
```

```python
# Line plot
plt.plot(df['Year'], df['Sales'])
plt.title('Sales Trend')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.show()
```
